```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your provided data
data = {
    'S no': [1, 2, 3, 4, 5, 6],
    'Country': ['China', 'India', 'United States', 'North Korea', 'Russia', 'Pakistan'],
    'Active Duty': [2185000, 1455550, 1388100, 1280000, 1014000, 654000],
    'Paramilitary': [660000, 2526950, 70000, 5889000, 554000, 291000],
    'Reserves': [1170000, 1155000, 844950, 600000, 2000000, 550000],
    'Total': [4015000, 5137500, 2233050, 7769000, 3568000, 1495000],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Displaying summary statistics
summary_statistics = df.describe()
print(summary_statistics)

# Visualizations

# Bar plot for Total military strength by country
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total', data=df, palette='coolwarm')
plt.title('Total Military Strength by Country')
plt.xticks(rotation=45)
plt.show()




# Trend of Active Duty over countries
plt.figure(figsize=(12, 6))
sns.lineplot(x='Country', y='Active Duty', data=df, marker='o')
plt.title('Trend of Active Duty Personnel by Country')
plt.xticks(rotation=45)
plt.show()




# Donut chart for the percentage of Paramilitary forces
plt.figure(figsize=(8, 8))
paramilitary_sizes = df['Paramilitary']
plt.pie(paramilitary_sizes, labels=df['Country'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Percentage of Paramilitary Forces by Country')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()




# Horizontal bar plot for Total military strength by country
plt.figure(figsize=(12, 8))
sns.barplot(x='Total', y='Country', data=df, palette='muted', orient='h')
plt.title('Total Military Strength by Country')
plt.show()
```
